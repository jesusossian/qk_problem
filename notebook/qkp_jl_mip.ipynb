```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_ = "100"
form = "linear"
```

```python
perc_ = "25"

table = pd.DataFrame()

file = f"../qkp_jl/result/{dir_}/{form}_{dir_}_{perc_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

tab = pd.DataFrame()
tab = pd.concat([tab, df], ignore_index=True)        
tab.columns = ['instance','form','lb','ub','gap','time','nodes','opt']

resume = pd.DataFrame({
    'instance':f'{perc_}',
    'form':f'{form}',
    'lb':tab['lb'].mean(),
    'ub':tab['ub'].mean(),
    "gap":tab.loc[tab["opt"] == 0,"gap"].mean(),
    "time":tab.loc[tab["opt"] == 1,"time"].mean(),
    'nodes':tab['nodes'].mean(),
    'opt':tab['opt'].sum(),
    }, index = ["resume"])

tab = pd.concat([tab, resume], ignore_index=True)

linear_25 = tab
linear_25
```
