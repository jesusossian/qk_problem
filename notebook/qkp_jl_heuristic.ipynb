```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_ = "100"
meth_ = "greedy"
```

```python
# std rf fo 
perc_ = "75"

file = f"../qkp_jl/result/{dir_}/{meth_}_{dir_}_{perc_}.txt"

df = pd.DataFrame()
df = pd.read_csv(file,header=None,sep=';')

table = pd.DataFrame()
table = pd.concat([table, df], ignore_index=True)
table.columns = ['instance','gre','ls','time']

resume = pd.DataFrame({
    'instance':f"{perc_}",
    'gre':table['gre'].mean(),
    'ls':table['ls'].mean(),
    'time':table['time'].mean()
    },index=["resume"]
)

table = pd.concat([table, resume], ignore_index=True)
table
```
